## _Filter() need more reading and practice._
The filter method selects columns by only inspecting the column names and not the actual data values. It has three mutually exclusive parameters, **items, like, and regex**, only one of which can be used at a time.   

**like**: takes a string and attempts to find all the column names that contain that exact string somewhere in the name.   

**regex**: instead to select column names through a regular expression. This particular regular expression, \d, represents all digits from zero to nine and matches any string with at least a single digit in it.  

**items**: which takes a list of exact column names. This is nearly an exact duplication of the indexing operator, except that a KeyError will not be raised if one of the strings does not match a column name. For instance, movie.filter(items=['actor_1_name', 'asdf']) runs without error and returns a single column DataFrame.

# Ordering column names sensibly

The following is a simple guideline to order columns:

* Classify each column as either discrete or continuous
* Group common columns within the discrete and continuous columns
* Place the most important groups of columns first with categorical columns before continuous ones

```python
movie = pd.read_csv('./data/movie.csv')
movie.head()
```

```python
movie.columns
```

```python
disc_core = [
    'movie_title', 
    'title_year',
    'content_rating',
    'genres'
]
disc_people = [
    'director_name', 
    'actor_1_name',
    'actor_2_name',
    'actor_3_name'
]
disc_other = [
    'color',
    'country',
    'language',
    'plot_keywords',
    'movie_imdb_link'
]
cont_fb = [
    'director_facebook_likes', 
    'actor_1_facebook_likes',
    'actor_2_facebook_likes',
    'actor_3_facebook_likes',
    'cast_total_facebook_likes',
    'movie_facebook_likes'
]
cont_finance = [
    'budget', 
    'gross'
]
cont_num_reviews = [
    'num_voted_users',
    'num_user_for_reviews',
    'num_critic_for_reviews']
cont_other = [
    'imdb_score',
    'duration',
    'aspect_ratio',
    'facenumber_in_poster'
]
```

```python
new_col_order = disc_core + \
disc_people + \
disc_other + \
cont_fb + \
cont_finance + \
cont_num_reviews + \
cont_other
set(movie.columns) == set(new_col_order)
```

```python
reordered_movie = movie[new_col_order]
reordered_movie.head()
```

A very common practice for relational databases is to have the primary key (if it exists) as the first column and any foreign keys directly following it.  
> Primary keys uniquely identify rows in the current table. Foreign keys uniquely identify rows in other tables.

# Working with operators on a DataFrame

When a DataFrame operates directly with one of the arithmetic or comparison operators, each value of each column gets the operation applied to it. Typically, when an operator is used with a DataFrame, the columns are either all numeric or all object (usually strings). If the DataFrame does not contain homogeneous data, then the operation is likely to fail.

```python
college = pd.read_csv('./data/college.csv')
college.head()
```

```python
# There columns represent the fraction of undergraduate students by race.
college = pd.read_csv('./data/college.csv', index_col='INSTNM')
college_ugds = college.filter(like='UGDS_')
college_ugds.head()
```

```python
(college_ugds + 0.00501).head()
```

```python
((college_ugds + 0.00501) // .01).head()
```

```python
college_ugds_op_round = ((college_ugds + 0.00501) // .01) / 100
college_ugds_op_round.head()
```

```python
college_ugds_round = (college_ugds + 0.00001).round(2)
```

```python
college_ugds_round.equals(college_ugds_op_round)
```

# Transposing the direction of a DataFrame operation

Many DataFrame methods have an axis parameter. This important parameter controls the direction in which the operation takes place. Axis parameters can only be one of two values, either 0 or 1, and are aliased respectively as the strings index and columns.  

Near all DataFrame methods default the axis parameter to 0/index.

```python
college_ugds.count()
# The count method returns the number of non-missing value
# By default, axis = 0
```

```python
college_ugds.count(axis=1).head()
```

```python
college_ugds.sum()
# sum method will return sum of each columns. Default axis = 0
```

```python
college_ugds.sum(axis='columns').head()
```

# Determining college campus diversity

```python
pd.read_csv('./data/college_diversity.csv', index_col='School')
```

```python
college_ugds.isnull().sum(axis=1).sort_values(ascending=False)
```

```python
college_ugds_dropedna = college_ugds.dropna(how='all')
college_ugds_dropedna.isnull().sum()
```

The dropna method has the how parameter, which is defaulted to the 'any' but may also be changed to 'all'. When set to any, it drops rows that contain one or more missing values. When set to all, it only drops rows where all values are missing.

```python
college_ugds_dropedna.ge(0.15).head()
```

```python
diversity_metric = college_ugds_dropedna.ge(0.15).sum(axis=1)
diversity_metric.head()
```

```python
diversity_metric.value_counts()
# There are two schools have more than 15% in five different race categories.
```

```python
diversity_metric.sort_values(ascending=False).head()
```
