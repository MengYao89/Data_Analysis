# Reducing memory by changing data types

```python
college = pd.read_csv('./data/college.csv')
cols_need = ['RELAFFIL', 'SATMTMID', 'CURROPER','INSTNM', 'STABBR']
col = college[cols_need].copy()
# college.loc[:, cols_need]
col.head()
```

```python
col.dtypes
```

```python
col.memory_usage()
```

```python
col.memory_usage(deep=True)
```

```python
col['RELAFFIL'] = col['RELAFFIL'].astype(np.int8)
```

```python
col.dtypes
```

```python
col.memory_usage(deep=True)
```

```python
col.select_dtypes(include=[np.object]).nunique()
```

```python
col['STABBR'] = col['STABBR'].astype('category')
```

```python
col.dtypes
```

```python
col.memory_usage(deep=True)
```

# Selecting the smallest of the latgest

```python
movie = pd.read_csv('./data/movie.csv')
need_col = ['movie_title', 'imdb_score', 'budget']
movie_2 = movie[need_col]
movie_2.head()
```

```python
movie_2.nlargest(50, 'imdb_score').head()
```

```python
movie_2.nlargest(50, 'imdb_score').nsmallest(5, 'budget')
```

# Selecting the larget of each group by sorting

```python
movie = pd.read_csv('./data/movie.csv')
movie_3 = movie[['movie_title', 'title_year', 'imdb_score']]
movie_3.head()
```

```python
movie_3.sort_values('title_year', ascending=False).head()
```

```python
movie_3.sort_values(['title_year', 'imdb_score'], ascending=False).head()
```

```python
movie_top_year = movie_3.drop_duplicates(subset='title_year')
movie_top_year.head()
# Default drop_duplicates will keep the first occurence of each unique row
# subset parameter only consider the column or list of columns
```

```python
movie_4 = movie[['movie_title', 'title_year','content_rating', 'budget']]
movie_4_sorted = movie_4.sort_values(['title_year', 'content_rating', 'budget'],
                                 ascending=[False, False, True])
movie_4_sorted.head()
```

```python
movie_4_sorted.drop_duplicates(subset=['title_year', 'content_rating']).head()
# can set keep parameter to last to select the last row of each group
```

# Replicating nlargest with sort_values

```python
movie = pd.read_csv('./data/movie.csv')
need_col = ['movie_title', 'imdb_score', 'budget']
movie_2 = movie[need_col].copy()
movie_2.head()
```

```python
movie_2.nlargest(10, 'imdb_score').nsmallest(5, 'budget')
```

```python
movie_2.dropna(inplace=True)
movie_2.sort_values('imdb_score', ascending=False).head(10).sort_values('budget').head(5)
```
