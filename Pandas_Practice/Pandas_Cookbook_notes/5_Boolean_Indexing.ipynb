# Calculating boolean statistics

```python
movie = pd.read_csv('./data/movie.csv', index_col='movie_title')
movie.head()
```

```python
movie_g_2_hours = movie['duration'] > 120
movie_g_2_hours.head()
```

```python
movie_g_2_hours.sum()
# There are 1039 movies duration longer than 2 hours.
```

```python
movie_g_2_hours.mean()
# Got the percentage of movies longer than 2 hours.
```

Due to the missing value in duration column, the output of above is misleading.

```python
movie[movie['duration'].isnull()]
```

```python
movie['duration'].dropna().gt(120).mean()
```

```python
movie_g_2_hours.describe()
```

```python
movie_g_2_hours.value_counts(normalize=True)
```

```python
np.nan > 120
# “Even though the duration column had missing values, 
# the boolean condition evaluated all these comparisons against 
# missing values as False.
```

# Constructing multiple boolean conditions

```python
movie = pd.read_csv('./data/movie.csv', index_col='movie_title')
```

```python
criteria_1 = movie.imdb_score > 8
criteria_2 = movie.content_rating == 'PG-13'
criteria_3 = (movie.title_year < 2000) | (movie.title_year > 2009)
```

```python
criteria_final = criteria_1 & criteria_2 & criteria_3
movie[criteria_final]
```
