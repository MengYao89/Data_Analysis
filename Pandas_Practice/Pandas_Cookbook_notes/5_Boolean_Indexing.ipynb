# Filtering with boolean indexing

```python
movie = pd.read_csv('./data/movie.csv', index_col='movie_title')
crit_a1 = movie['imdb_score'] > 8
crit_a2 = movie['content_rating'] == 'PG-13'
crit_a3 = (movie['title_year'] < 2000) | (movie['title_year'] > 2009)
final_crit_a = crit_a1 & crit_a2 & crit_a3
```

```python
crit_b1 = movie['imdb_score'] < 5
crit_b2 = movie['content_rating'] == 'R'
crit_b3 = (movie['title_year'] >= 2000) & (movie['title_year'] <= 2010)
final_crit_b = crit_b1 & crit_b2 & crit_b3
```

```python
final_crit_all = final_crit_a | final_crit_b
final_crit_all.head()
```

```python
movie[final_crit_all].head()
```

```python
cols = ['imdb_score', 'content_rating', 'title_year']
movie_filtered = movie.loc[final_crit_all, cols]
movie_filtered.head(8)
```

# Replicating boolean indexing with index selection

```python
college = pd.read_csv('./data/college.csv')
college[college['STABBR'] == 'TX'].head()
```

```python
college_2 = college.set_index('STABBR')
college_2.loc['TX'].head()
```

```python
%timeit college[college['STABBR'] == 'TX']
```

```python
%timeit college_2.loc['TX']
```

```python
%timeit college_2 = college.set_index('STABBR')
```

As we can see from above that Boolean indexing takes three times as long as index selection, but setting index also not come for free.

# Selecting with unique and sorted indexes

Index selection performance drastically improves when the index is unique or sorted. We can form unique or sorted indexes to increase the performance of index selection.

```python
college = pd.read_csv('./data/college.csv')
college2 = college.set_index('STABBR')
college2.index.is_monotonic
```

```python
college3 = college2.sort_index()
college3.index.is_monotonic
```

```python
%timeit college[college['STABBR'] == 'TX']
```

```python
%timeit college2.loc['TX']
```

```python
%timeit college3.loc['TX']
```

The sorted index performs nearly an order of magnitude faster than boolean selection.

```python
college_unique = college.set_index('INSTNM')
college_unique.index.is_unique
```

```python
%timeit college[college['INSTNM'] == 'Stanford University']
```

```python
%timeit college_unique.loc['Stanford University']
```

# Translating SQL WHERE clauses

Many pandas users will have a background processing data directly from databases using the ubiquitous Structured Query Language (SQL). SQL is a standardized language to define, manipulate, and control data stored in a database. The SELECT statement is the most common way to use SQL to select, filter, aggregate, and order data. Pandas has the ability to connect to databases and send SQL statements to them.

> Example:  
> find all the female employee that work in the ploce or fire departments that have a base salary between 80 and 120 thousand dollars.  

SQL:   
SELECT  
    UNIQUE_ID,  
    DEPARTMENT,  
    GENDER,  
    BASE_SALARY  
FROM  
    EMPLOYEE  
WHERE  
    DEPARTMENT IN ('Houston Polica Department-HPD, 'Houston Fire Department (HFD)')
    AND  
    GENDER = 'Female'  
    AND
    BASE_SALARY BETWEEN 80000 AND 120000

```python
employee = pd.read_csv('./data/employee.csv')
```

The Series **value_counts** method is an excellent way to get both the exact string name and number of occurrences of that value.

```python
employee['DEPARTMENT'].value_counts().head()
```

```python
employee['GENDER'].value_counts().head()
```

```python
employee['BASE_SALARY'].describe().astype(int)
```

```python
dept = ['Houston Police Department-HPD', 'Houston Fire Department (HFD)']
criteria_dept = employee['DEPARTMENT'].isin(dept)
criteria_gender = employee['GENDER'] == 'Female'
criteria_salary = (employee['BASE_SALARY'] >= 80000) & (employee['BASE_SALARY'] <= 120000)
# criteria_salary = employee['BASE_SALARY].between(80000, 120000)

criteria_final = (criteria_dept & criteria_gender & criteria_salary)
```

The **isin** Series method is equivalent to the SQL IN operator and accepts a list of all possible values that you would like to keep.

```python
select_col = ['UNIQUE_ID', 'DEPARTMENT', 'GENDER', 'BASE_SALARY']
employee.loc[criteria_final, select_col].head().reset_index(drop=True)
```
