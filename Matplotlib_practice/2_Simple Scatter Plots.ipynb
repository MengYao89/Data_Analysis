```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0,10,20)
y = np.sin(x)

plt.plot(x, y, linestyle='--', marker='o', color='orange')
# Third args 'o' mean scatter plots
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{}'".format(marker))
    plt.legend()
    plt.xlim(0, 1.5)
```

```python
plt.plot(x, y, color='orange',
         linewidth=4,
         marker='o',
         markersize=14,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2         
)
plt.ylim(-2, 2)

# color define the line color
# markersize define the marker size
# markerfacecolor define the inner marker color
# markeredgecolor define the marker outter line color
# markeredgewidth define the marker outer line width
```

```python
plt.scatter(x, y, marker='o', color='red')
```

```python
rng = np.random.RandomState(0)

x = rng.randn(50)
y = rng.randn(50)
colors = rng.rand(50)
sizes = 1000 * rng.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
# alpha is transparency level
plt.colorbar(); # show color scale
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.3,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);

iris.feature_names
```
