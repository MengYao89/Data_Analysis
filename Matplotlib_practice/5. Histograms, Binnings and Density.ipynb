```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
data = np.random.randn(1000)
```

```python
plt.hist(data);
```

```python
plt.hist(data, 
         bins=30, 
         normed=True, 
         alpha=0.5, 
         histtype='stepfilled', # {‘bar’, ‘barstacked’, ‘step’, ‘stepfilled’}
         color='steelblue',
         edgecolor='red' # or None
);
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.4, normed=True, bins=60)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
# count the number of points in a given bin
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

## Two-Dimensional Histograms and Binnings

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]

x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
plt.colorbar().set_label('Counts in bin')
```

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
plt.colorbar().set_label('Counts in bin')
```
