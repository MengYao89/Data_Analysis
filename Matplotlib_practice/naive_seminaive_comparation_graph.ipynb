```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 3
naive_value = (0.076, 9.373, 126.419)
seminaive_value = (0.062, 4.869, 92.307)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots(figsize=(10,8))
rects1 = ax.bar(ind, naive_value, width, color='blue')
rects2 = ax.bar(ind + width, seminaive_value, width, color='orange')

# add some text for labels, title and axes ticks
ax.set_ylabel('Excution Time', fontsize=14)
ax.set_title('Clique comparation of Naive and Semi-Naive', fontsize=16)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Clique10', 'Clique100', 'Clique200'), fontsize=14)

ax.legend((rects1[0], rects2[0]), ('Naive', 'Semi-Naive'), fontsize=12)

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                '%0.3f s' % height,
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```
